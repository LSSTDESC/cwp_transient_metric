```python
def num_obs_bands(lc, SNR_thresh=1.0):
    """
    Return a dictionary of number of observations in each bandpass
    
    Parameters:
    -----------
    lc: instance of LightCurve
        light curve object with information available
    SNR_threshold: float, defaults to 1.0
        Evaluate the number of observations when the SNR is above the threshold.
    """
    df = lc.lightCurve
    df['SNR'] = df.flux / df.fluxerr
    return df.query('SNR > @SNR_thresh').groupby('band').agg(dict(SNR='count'))
```
